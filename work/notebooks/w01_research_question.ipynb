## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

### Decision

The decision is to identify which content pages should be reviewed first for refresh or improvement.

### Who acts?

The output will be used by content reviewers, SEO specialists, or content managers.

### Action

The reviewer will prioritize the highest-ranked pages for manual review and decide whether they should be refreshed, expanded, rewritten, monitored, or left unchanged.

### Cost of a wrong recommendation

If a low-priority page is recommended, reviewers may waste valuable time reviewing pages that do not need immediate attention.

If an important page with declining performance is missed, valuable traffic and business opportunities may be lost.

### Why ML?

The dataset contains many observable signals such as impressions, clicks, CTR, engagement rate, content age, and freshness. These signals interact in complex ways that are difficult to capture with simple rules. Machine learning can help prioritize pages while still supporting human decision-making.
